## 1.2 sleep equation #1

```python
data_string = '''R2                      0.115          0.126    '''
```

```python
R_res,R_unres = np.float_(re.findall(r'([-+]?\d+.\d+)', data_string))
```

$$ H_0: \beta_{male} = \beta_{totwrk:male} = \beta_{age:male} = \beta_{smsa:male} = 0$$
$$ H_1: \beta_{male}^2 + \beta_{totwrk:male}^2 + \beta_{age:male}^2 + \beta_{smsa:male}^2 >0$$

```python
F_nabl = (R_unres-R_res)/(1-R_unres)*(len(df)-8)/4
F_nabl.round(2)
```

```python
alpha = 0.05
stats.f.ppf(1-alpha,4,len(df)-8).round(2)
```

$F_{nabl}<F_{crit} =>$ данные согласуются с гипотезой $Н_0 =>$ структурные сдвиги незначимы

# 2 Небинарные качественые признаки

## 2.1 diamond equation #1

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/Diamond.csv')
df
```

```python
df_all = pd.get_dummies(df.loc[:,['carat','colour','certification','price']])
df_all
```

```python
model_d = smf.ols(formula='np.log(price)~carat+I(carat**2)+colour_E+colour_F+colour_G+colour_H+colour_I+certification_HRD+certification_IGI',data=df_all).fit()
model_d.params
```

$$ H_0: \beta_{colour_E} = \beta_{colour_F} = \beta_{colour_G} = \beta_{colour_H} = \beta_{colour_I} = 0$$
$$ H_1: \beta_{colour_E}^2 + \beta_{colour_F}^2 + \beta_{colour_G}^2 + \beta_{colour_H}^2 + \beta_{colour_I}^2 > 0$$

```python
model_d.f_test('colour_E=colour_F=colour_G=colour_H=colour_I=0')
```

```python
alpha=0.01
stats.f.ppf(1-alpha,5,len(df_all)-10).round(2)
```

$F_{nabl}>F_{crit} =>$ данные не согласуются с гипотезой $Н_0 =>$ влияние значимо

$$ H_0: \beta_{certification_{HRD}} = \beta_{certification_{IGI}} = 0$$
$$ H_1: \beta_{certification_{HRD}}^2 + \beta_{certification_{IGI}}^2 > 0$$

```python
model_d.f_test('certification_HRD=certification_IGI=0')
```

```python
alpha=0.01
stats.f.ppf(1-alpha,2,len(df_all)-8).round(2)
```

$F_{nabl}>F_{crit} =>$ данные не согласуются с гипотезой $Н_0 =>$ влияние значимо

## 2.2 diamond equation #2

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/diamonds.csv')
df
```

```python
df_all = pd.get_dummies(df.loc[:,['price','carat','color','cut','x','y','z' ]])
df_all
```

```python
df_all.columns = df_all.columns.str.replace('cut_Very Good', 'cut_VeryGood')
df_all
```

```python
model_d = smf.ols(formula='np.log(price)~carat+I(carat**2)+color_E+color_F+color_G+color_H+color_I+color_J+cut_Good+cut_VeryGood+cut_Premium+cut_Ideal+x+y+z',data=df_all).fit()
model_d.params
```

$$ H_0: \beta_{color_E} = \beta_{color_F} = \beta_{color_G} = \beta_{color_H} = \beta_{color_I} = \beta_{color_J}= 0$$
$$ H_1: \beta_{color_E}^2 + \beta_{color_F}^2 + \beta_{color_G}^2 + \beta_{color_H}^2 + \beta_{color_I}^2 + \beta_{color_J}^2 > 0$$

```python
model_d.f_test('color_E=color_F=color_G=color_H=color_I=color_J=0')
```

```python
alpha=0.05
stats.f.ppf(1-alpha,6,len(df_all)-16).round(2)
```

$F_{nabl}>F_{crit} =>$ данные не согласуются с гипотезой $Н_0 =>$ влияние значимо

$$ H_0: \beta_{cut_{Good}} = \beta_{cut_{VeryGood}} = \beta_{cut_{Premium}} = \beta_{cut_{Ideal}} = 0$$
$$ H_1: \beta_{cut_{Good}}^2 + \beta_{cut_{VeryGood}}^2 + \beta_{cut_{Premium}}^2 + \beta_{cut_{Ideal}}^2 > 0$$

```python
model_d.f_test('cut_Good=cut_VeryGood=cut_Premium=cut_Ideal=0')
```

```python
alpha=0.05
stats.f.ppf(1-alpha,4,len(df_all)-16).round(2)
```

$F_{nabl}>F_{crit} =>$ данные не согласуются с гипотезой $Н_0 =>$ влияние значимо
